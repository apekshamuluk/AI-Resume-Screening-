### 1] Import Required Libraries.

```python
import PyPDF2        # reads PDF
import nltk          # text cleaning

import os
import sys
sys.path.append(os.path.abspath('.'))

from sklearn.feature_extraction.text import TfidfVectorizer         # converts text into numbers
from sklearn.metrics.pairwise import cosine_similarity              # compares text

import re
nltk.download('stopwords')
from nltk.corpus import stopwords
```

### 2] Import Modules And Set Path

```python
from src.parser import extract_text
from src.preprocess import preprocess
from src.vectorizer import vectorize
from src.matcher import compute_similarity
from src.ranker import rank_resumes

# MAIN FUNCTION: ANALYZE AND RANK RESUMES

def analyze_resumes(uploaded_files, job_desc):
    
    resumes = []
    
    for file in uploaded_files:
        text = extract_text(file)
        clean_text = preprocess(text)
        resumes.append(clean_text)
    
    job_desc_clean = preprocess(job_desc)
    
    all_text = resumes + [job_desc_clean]
    
    vectors = vectorize(all_text)
    
    resume_vectors = vectors[:-1]
    job_vector = vectors[-1]
    
    scores = compute_similarity(resume_vectors, job_vector)
    
    result = rank_resumes(scores)
    
    return result
```

### 3] Input Data (Resumes + Job Description)

```python
uploaded_files = [
    r"C:\Users\Apeksha Muluk\Downloads\Suraj_Jawale.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\Tejal_Padwal_.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\Apeksha_Muluk_Resume.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\Rohan_Gaikwad.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\Sushil Ambekar resume.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\VAISHNAVI DUMBRE_RESUME.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\RutujaMunde_resume.pdf",
    r"C:\Users\Apeksha Muluk\Downloads\ghodake_siddhi.pdf"
]

job_desc = """Python SQL Excel Power BI Data Analysis Data Cleaning Data Visualization Machine Learning Statistics Pandas NumPy Matplotlib 
            Scikit-learn Dashboard Data Analysis Data Cleaning Data Visualization Python SQL Excel Power BI Pandas NumPy Matplotlib Scikit-learn 
            Machine Learning Statistics Dashboard Data Analysis Python SQL Excel Power BI Data Analysis Data Cleaning Data Visualization Machine 
            Learning Pandas NumPy Matplotlib Scikit-learn"""
```

### 4] Output: Display Ranked Results

```python
result = analyze_resumes(uploaded_files, job_desc)

for rank, (index, score) in enumerate(result):
    name = os.path.basename(uploaded_files[index])
    
    percentage = score * 100
    status = "Selected ✅" if score > 0.25 else "Rejected ❌"
    
    print(f"Rank {rank+1}: {name}")
    print(f"Score: {score:.2f} ({percentage:.1f}%) → {status}")
    print("-" * 40)

# Top candidate
top_index, top_score = result[0]
top_name = os.path.basename(uploaded_files[top_index])

print(f"\n🏆 Best Candidate: {top_name} ({top_score:.2f})")
```
